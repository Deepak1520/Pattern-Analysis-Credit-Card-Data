# Couting the no. of Deliquent and Non-Deliquent Customers and Visualizing it

```python


delinquent = df['TARGET_LABEL'].sum() #add all the 1's the target_level column
non_delinquent = df.shape[0] - delinquent #subtracting the total number of rows and the no. of deliquent customers
delinquent_percent = (delinquent/df.shape[0])*100
non_delinquent_percent = (non_delinquent/df.shape[0])*100


print("Number of delinquent customers: {}({:.2f}%)".format(delinquent, delinquent_percent))
print("Number of non-delinquent customers: {}({:.2f}%)".format(non_delinquent, non_delinquent_percent))
```

```python
#From the above analysis of the provided dataset, 
#it can be observed that the majority of customers have a good credit history and are non-delinquent, 
#while only a small percentage of customers have a history of delinquency.
```

```python
#To support this observation, we can visualize the distribution of 
#delinquent and non-delinquent customers using a bar chart.

import seaborn as sns
import matplotlib.pyplot as plt


label_dict = {0:"Non-Deliquent", 1:"Deliquent"}

#plt.figure(figsize = (8,8))
sns.countplot(x='TARGET_LABEL', data=df, hue = "TARGET_LABEL")
plt.legend(labels=label_dict.values())

plt.title("Distribution of Deliquent and Non-Deliquent customers")
plt.show()
```

# Exploring the relationship between the target variable and gender in the dataset.


```python
import pandas as pd

# group the dataset by GENDER and TARGET_LABEL and count the number of instances in each group
gender_counts = df.groupby(['GENDER', 'TARGET_LABEL']).size().reset_index(name='count')

# filter the groups to keep only male and female customers
gender_counts = gender_counts.loc[gender_counts['GENDER'].isin(['M', 'F'])]

# replace TARGET_LABEL values with 'Delinquent' and 'Non-Delinquent'
gender_counts['TARGET_LABEL'] = gender_counts['TARGET_LABEL'].replace({0: 'Non-Delinquent', 1: 'Delinquent'})

# pivot the table to put TARGET_LABEL values into separate columns
gender_counts = gender_counts.pivot(index='GENDER', columns='TARGET_LABEL', values='count')

# calculate the percentage of delinquent and non-delinquent customers for each gender
gender_counts['Delinquent %'] = (gender_counts['Delinquent'] / (gender_counts['Delinquent'] + gender_counts['Non-Delinquent'])) * 100
gender_counts['Non-Delinquent %'] = (gender_counts['Non-Delinquent'] / (gender_counts['Delinquent'] + gender_counts['Non-Delinquent'])) * 100

# display the resulting table
print(gender_counts)
```

```python
# from above we can conclude that male are more likely to be delinquent than female
```

```python
# Map TARGET_LABEL values to delinquent and non-delinquent
df['TARGET_LABEL'] = df['TARGET_LABEL'].map({0: 'Non-Delinquent', 1: 'Delinquent'})

# group the dataset by GENDER and TARGET_LABEL and count the number of instances in each group
gender_counts = df.groupby(['GENDER', 'TARGET_LABEL']).size().reset_index(name='count')

# filter the groups to keep only male and female customers
gender_counts = gender_counts.loc[gender_counts['GENDER'].isin(['M', 'F'])]

# pivot the table to put TARGET_LABEL values into separate columns
plt.figure(figsize = (15,8))
gender_counts = gender_counts.pivot(index='GENDER', columns='TARGET_LABEL', values='count')

# plot the stacked bar chart
gender_counts.plot(kind='bar', stacked=True)

# set the title and axis labels
plt.title('Delinquent vs Non-Delinquent Customers by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')

# show the plot
plt.show()
```
